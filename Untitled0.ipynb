# Titanic Survival Analysis (Python EDA)

This project analyzes passenger data from the Titanic disaster to identify key factors that influenced survival rates, including class, gender, age, and fare.

## Objective
The goal of this project is to explore the Titanic dataset and uncover patterns that influenced passenger survival using Python and data visualization techniques.

## Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head()
```

```python
survival_by_gender = df.groupby('sex')['survived'].mean()

plt.figure()
survival_by_gender.plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
### Insight: Female passengers had a significantly higher survival rate than male passengers, indicating that evacuation protocols prioritized women during the disaster.
```

```python
# Create age groups
df['age_group'] = pd.cut(df['age'], bins=[0, 12, 18, 35, 60, 100],
                         labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

# Survival by age group
survival_by_age = df.groupby('age_group')['survived'].mean()

# Plot
plt.figure()
survival_by_age.plot(kind='bar')
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
### Insight: Children had higher survival rates compared to adults, supporting the “women and children first” evacuation protocol, while older passengers had lower survival rates.
```

```python
# Scatter plot of fare vs survival
plt.figure()

plt.scatter(df['fare'], df['survived'], alpha=0.5)

plt.title("Fare vs Survival")
plt.xlabel("Fare Paid")
plt.ylabel("Survived (0 = No, 1 = Yes)")

plt.tight_layout()
plt.show()
### Insight: Passengers who paid higher fares were more likely to survive, suggesting that wealthier individuals had better access to safety and lifeboats during the disaster.
```
